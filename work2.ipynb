```python
# How many student who got lowest maarks in CT-3

len(new_file[new_file["CT-3"] == new_file["CT-3"].min()])
```

```python
# How many student who got above average marks in CT-2

average_marks = new_file["CT-2"].mean()
above_average_students = len(new_file[new_file["CT-2"] > average_marks])
print(f"Number of students who got above average marks in CT-2: {above_average_students}")
```

```python
# Add Total Column to mar dataframe

new_file["Total"] = new_file["CT-1"] + new_file["CT-2"] + new_file["CT-3"]
new_file.head()
```

```python
# Update Total column with sum of CT-1,CT-2 and CT-3

new_file["Total"] = new_file["CT-1"] + new_file["CT-2"] + new_file["CT-3"]
new_file.head()
```

```python
# Add Percentage Column to mar dataframe and update this column Total/90

new_file["Percentage"] = round(new_file["Total"] / 90 * 100, 2)
new_file.head()
```

```python
# How many students got more than 75% marks

students_above_75 = len(new_file[new_file["Percentage"] > 75])
print(f"Number of students who got more than 75% marks: {students_above_75}")
```

```python
# How many students appear only in 2 sessionals

students_in_two_sessionals = len(new_file[new_file["CT-1"] == 0][new_file["CT-2"] != 0][new_file["CT-3"] == 0]) + len(new_file[new_file["CT-1"] != 0][new_file["CT-2"] == 0][new_file["CT-3"] == 0]) + len(new_file[new_file["CT-1"] == 0][new_file["CT-2"] == 0][new_file["CT-3"] != 0])
print(f"Number of students who appeared in only 2 sessionals: {students_in_two_sessionals}")
```

```python
# How many students appear at least in 2 sessionals

students_in_at_least_two_sessionals = len(new_file[(new_file["CT-1"] != 0) & (new_file["CT-2"] != 0)]) + len(new_file[(new_file["CT-1"] != 0) & (new_file["CT-3"] != 0)]) + len(new_file[(new_file["CT-2"] != 0) & (new_file["CT-3"] != 0)])
print(f"Number of students who appeared in at least 2 sessionals: {students_in_at_least_two_sessionals}")
```

```python
# How many students appear at most in 2 sessionals

students_in_at_most_two_sessionals = len(new_file[(new_file["CT-1"] == 0) & (new_file["CT-2"] == 0)]) + len(new_file[(new_file["CT-1"] == 0) & (new_file["CT-3"] == 0)]) + len(new_file[(new_file["CT-2"] == 0) & (new_file["CT-3"] == 0)]) + students_in_at_least_two_sessionals
print(f"Number of students who appeared in at most 2 sessionals: {students_in_at_most_two_sessionals}")
```
