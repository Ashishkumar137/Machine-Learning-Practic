```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
plt.scatter(tips["total_bill"],tips["tip"])
plt.xlabel("Total bill")
plt.ylabel("Tip")
plt.title("Total bill vs Tip")
plt.show()
```

```python
# Write Python code to calculate the mean, median, and standard deviation of a numerical variable in a pandas DataFrame.
```

```python
print("Mean =" , tips["total_bill"].mean())
print("Median =" , tips["total_bill"].median())
print("Standerd Deviation =" , tips["total_bill"].std())
```

```python
tips.describe()
```

```python
# Write Python code to create a histogram using Seaborn for visualizing the distribution of a numerical variable from a pandas DataFrame.
```

```python
sns.histplot(tips["total_bill"])
plt.xlabel("Total bill")
plt.ylabel("Count")
```

```python
# Write Python code to calculate the correlation coefficient between two numerical variables in a pandas DataFrame.
```

```python
tips["total_bill"].corr(tips["tip"])
```

```python
# Write Python code to create a box plot using Matplotlib for visualizing the distribution of a numerical variable across different categories from a pandas DataFrame.
```

```python
tips.boxplot(column= "total_bill", by = "day")
plt.xlabel("Total bill")
plt.ylabel("Tip")
plt.title("Total bill vs Tip")
plt.show()
```
